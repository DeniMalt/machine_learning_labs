В лабораторной работе №1 необходимо решить 15 заданий на языке программирования python.
  1. Импортируйте необходимые библиотеки
  2. Импортируйте набор данных из файла cafe.tsv и поместите его в
  объект DataFrame
  3. Выведите 12 первых строк
  4. Сколько строк и столбцов в наборе данных?
  5. Получите сводную информацию по датафрейму
  6. Что является индексом набора данных?
  7. Сколько уникальных товаров упоминается в наборе данных?

  8. Определите количество проданных товаров каждого наименова-
  ния, отсортируйте по убыванию и найдите пять товаров, пользую-
  щихся наибольшим спросом

  9. Какие из описаний choice_description упоминаются в заказах чаще
  всего?
  10. Каково общее количество купленных товаров?
  11. Сколько всего заказов содержит набор данных?
  12. Преобразуйте значения item_price в числа с плавающей запятой
  13. Каков доход кафе за рассматриваемый период?
  14. Какова средняя цена заказа?
  15. Получите сводку статистик для числового столбца item_price

Решение:

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/cafe.tsv', sep='\t', skiprows=0)

print(df.head(12))

print(f"\n4. Размер набора данных: {df.shape[0]} строк, {df.shape[1]} столбцов")

print("\n5. Сводная информация:")
print(df.info())

print(f"\n6. Индекс набора данных: {df.index}")

unique_items = df['item_name'].nunique()
print(f"\n7. Количество уникальных товаров: {unique_items}")

item_sales = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
top_5_items = item_sales.head(5)
print("\n8. Пять самых популярных товаров:")
print(top_5_items)

choice_counts = df['choice_description'].value_counts().head(10)
print("\n9. Самые частые описания choice_description:")
print(choice_counts)

total_items = df['quantity'].sum()
print(f"\n10. Общее количество купленных товаров: {total_items}")

total_orders = df['order_id'].nunique()
print(f"\n11. Количество уникальных заказов: {total_orders}")

df['item_price'] = df['item_price'].str.replace('$', '').astype(float)
print("\n12. Цены преобразованы в числовой формат")
print(f"Тип данных столбца item_price: {df['item_price'].dtype}")

df['total_price'] = df['quantity'] * df['item_price']
total_revenue = df['total_price'].sum()
print(f"\n13. Общий доход кафе: ${total_revenue:.2f}")

order_totals = df.groupby('order_id')['total_price'].sum()
average_order_value = order_totals.mean()
print(f"\n14. Средняя стоимость заказа: ${average_order_value:.2f}")

print("\n15. Статистика цен товаров:")
price_stats = df['item_price'].describe()
print(price_stats)
```

Таким образом, в лабораторной работе №1 были выполнены практические задачи, связанные с анализом данных, с помощью библиотеки pandas.
